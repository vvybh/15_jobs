```python
# 자료형의 참과 거짓 
print(bool("python"))
print(bool(""))
print("="*10)
print(bool([1, 2, 3]))
print(bool([]))
print("="*10)
print(bool(5))
print(bool(0))
print("="*10)
print(bool({}))
print(bool(set()))
```

```python
# 논리연산자 
# and 연산자

print(True and True)
print(True and False)
print(False and True)
print(False and False)

print("="*10)

# or 연산자
print(True or True)
print(True or False)
print(False or True)
print(False or False)

print("="*10)
# not 연산자
print(not True)
print(not False)
print(not 1)
print(not 0)
```
