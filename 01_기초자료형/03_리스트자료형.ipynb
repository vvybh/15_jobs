```python
# 리스트의 수정과 삭제 
a = [1, 2, 3]
a[2] = 4
print(a)
```

```python
# 리스트 관련 함수 

# 1. 리스트에 요소 추가하기 - append 
a = [1, 2, 3]
a.append(4)
print(a)
```

```python
a.append([5, 6])
print(a)
```

```python
# 2. 리스트 정렬 - sort 
a = [ 5, 1, 4, 3, 2]
a.sort()
print(a)
```

```python
a.sort(reverse=True)
print(a)
```

```python
# 3. 리스트 뒤집기 - reverse 
a = [ 5, 1, 4, 3, 2]
a.reverse()
print(a)
```

```python
# 4. 인덱스 반환 - index
a = [1, 2, 3]
a.index(3)
```

```python
# 5. 리스트에 요소 삽입 - insert
a = [1, 2, 3]
a.insert(0, 4)
print(a)
```

```python
# 6. 리스트 요소 제거 - remomve
a = [1, 2, 3, 1, 2, 3]
a.remove(3)
print(a)
```

```python
# 7. 리스트 요소 제거하기 - pop
a = [1, 2, 3]
a.pop()
```

```python
# 8. 리스트에 포함된 요소 x의 개수 세기 - count 
a = [1, 2, 3, 1]
a.count(1)
```

```python
# 9. 리스트 확장 - extend

a = [1, 2, 3]
a.extend([4, 5])
print(a)
```
