```python
# 숫자형은 어떻게 사용할까? 

## 정수형

a = 123
b = -11
c = 0 

print(a)
print(b)
print(c)

print(type(a))
print(type(b))
print(type(c))


```

```python
## 실수형
a = 1.2 
b = -3.14
c = 4.2E5
d = 4.2E-5

print(a)
print(b)
print(c)
print(d)

print(type(a))
print(type(b))
print(type(c))
print(type(d))
```

```python
## 8진수 16진수 
# 314
# 3*(10**2) + 1*(10**1) + 4

a = 0o177
print(a)
print(1 * (8**2) + 7 * (8**1) + 7)
b = 0o10
c = 0o7

print(b)
print(c)

```

```python
a = 0x8ff 
b = 0xabc

# 123456789abcdef
print(a)
print(8 * (16 ** 2) + 15 * (16 ** 1) + 15)

print(10 * (16 ** 2) + 11 * (16 ** 1) + 12)
print(b)
```

```python
# 사칙연산
a = 3
b = 4

print(a + b)
print(a - b)
print(a * b)
print(a / b)
```

```python
# x의 y 제곱을 나타내는 ** 연산자

a = 3
b = 4

print(a ** b)
```

```python
# 나머지를 구하는 % 연산자
7 % 3 
```

```python
# 나눗셈 후 몫을 반환하는 // 연산자
7 // 4
```

```python
# 복합연산자
a = 1 
a += 1 # a = a + 1

# +=
# -=
# *=
# /=
# %=
# **=
# //=

print(a)
```
