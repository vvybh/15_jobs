```python
# 딕셔너리에서 key를 사용하여 value값 얻기

data = {
    "유재석" : "개그맨", 
    "손흥민" : "축구선수", 
    "귀도": "개발자"
}

print(type(data))

print(data["유재석"])
print(data["손흥민"])

data["귀도"] = "코딩"
print(data)

data["강호동"] = "개그맨"
print(data)
```

```python
# 딕셔너리를 만들때 주의사항 
# 1. 딕셔너리에서 key는 고유한 값이므로 중복이 안됨
# 2. 수정이 되는 데이터 타입을 key 로 올 수 없다.(리스트)

a = {
    [1, 2] : "hi"
}
print(a)
```

```python
a = {
    1: "a", 
    "a" : 10
}
print(a)
```

```python
a = 1
print(a)
```

```python
a = 10
print(a)
```

```python
# 딕셔너리 관련함수 
# key 리스트 만들기 - keys

data = {
    "유재석" : "개그맨", 
    "손흥민" : "축구선수", 
    "귀도": "개발자"
}

print(list(data.keys()))
```

```python
# Value 리스트 만들기 - values
data = {
    "유재석" : "개그맨", 
    "손흥민" : "축구선수", 
    "귀도": "개발자"
}

print(list(data.values()))
```

```python
# key, values 쌍 얻기 - items
data = {
    "유재석" : "개그맨", 
    "손흥민" : "축구선수", 
    "귀도": "개발자"
}
print(list(data.items()))
```

```python
# key, value 쌍 모두 지우기 - clear 
data.clear() 
print(data)
```

```python
# key로 value 값 얻디 - get 
data = {
    "유재석" : "개그맨", 
    "손흥민" : "축구선수", 
    "귀도": "개발자"
}

print(data.get("유재석"))
print(data.get("김연아"))
print(data["김연아"])
```

```python
# 해당 key가 딕셔너리 안에 있는지 조사하기 - in
data = {
    "유재석" : "개그맨", 
    "손흥민" : "축구선수", 
    "귀도": "개발자"
}

print("유재석" in data) 
print("김연아" in data) 
```

```python
# key로 value 꺼내기 - pop 
data = {
    "유재석" : "개그맨", 
    "손흥민" : "축구선수", 
    "귀도": "개발자"
}
poped = data.pop("유재석")
print(poped)
print(data)
```
