```python
from faker import Faker 


fake = Faker("ko-kr")

fake.name()
```

```python
fake.address()
```

```python
lis = []
for _ in range(10): 
    lis.append({
        "이름": fake.name(), 
        "주소" : fake.address(), 
        "국가명": fake.country(), 
        "회사명": fake.company(), 
        "직업": fake.job(), 
        "전화번호": fake.phone_number()
        })

print(lis)
```

```python
with open("profiles.csv", "w") as file: 
    file.write("이름, 전화번호, 직업, 회사명, 주소\n")
    for _ in range(10): 
        name = fake.name()
        phone = fake.phone_number()
        job = fake.job()
        company = fake.company()
        address = fake.address()
        file.write(f"{name}, {phone}, {job}, {company}, {address}\n")
```
