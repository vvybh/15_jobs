```python
# for문의 기본 구조 

# for 변수 in 리스트(또는 튜플, 문자열):
#     수행할 문장1
#     수행할 문장2
#     ...
```

```python
test_list = ["one", "two", "three"]

for li in test_list:
    print(li)
```

```python
# 다양한 for문의 사용
a = [(1, 2), (3, 4), (5, 6)]
for i, j in a:
    print(f"{i} X {j} = {i*j}")
```

```python
marks = [90, 25, 70, 45, 80]
number = 0

for mark in marks:
    number = number + 1
    if mark >= 60: 
        print(f"{number}번 학생은 합격입니다.")
    else: 
        print(f"{number}번 학생은 불합격입니다.")
```

```python
# for문 continue 문

marks = [90, 25, 70, 45, 80]
number = 0

for mark in marks: 
    number += 1 # number = number + 1
    if mark < 60: 
        continue

    print(f"{number}번 학생은 합격입니다.")
```

```python
# for문과 함께 자주 사용하는 range함수 

a = range(10)
print(a)
print(list(a))
```

```python
b = range(1, 11) 
print(list(b))
```

```python
add = 0 

for number in range(1, 10+1):
    add = add + number 
    print(add)

print(add)
```

```python
# for와 range 합수를 이용해서 구구단 
for i in range(2, 10): 
    for j in range(1, 10): 
        print(f"{i} X {j} = {i * j}")
```

```python
# *
# **
# ***
# ****
# *****

for i in range(1, 6): 
    print("*"*i)

#     *
#    **
#   ***
#  ****
# *****

for i in range(1, 6):
    print(" " * (5-i) + "*"*i)
```

```python
# 리스트 컴프리헨션 사용하기 

a = [1, 2, 3, 4]
result = []
for num in a: 
    result.append(num*3)

print(result)
```

```python
a = [1, 2, 3, 4]
result = [num * 3 for num in a]
print(result)
```

```python
# for문과 break문
for i in range(10): 
    if i == 3: 
        break
    print(i)
```

```python
# enumerate 함수 사용하기 

fruits = ["사과", "바나나", "오렌지"]
for number, fruit in enumerate(fruits): 
    print(f"{number}번째 과일은 {fruit}이다")
```

```python
# zip함수로 여러 리스트 함께 순회하기 
names = ["홍길동", "유재석", "강호동"]
scores = [70, 80, 90]


for (name, score) in zip(names, scores): 
    print(f"{name}의 점수는 {score}점이다")
```
